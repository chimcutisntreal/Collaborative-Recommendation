<h3>Bước 1:</h3> 
Đầu vào là Tập bài báo (Mỗi hàng sẽ là một bài báo, như vậy ta có một bảng với số hàng là số bài báo, các cột là các thông tin đi kèm của bài báo đó

```python
# ta chỉ lấy 10 nghìn bài báo từ tập dữ liệu
df = df_1[:][:10000]
```

```python
# lọc ra cột author lưu vào biến df_author
df_author = df[['author']]
df_author.head()
```

```python
# Lấy danh sách các tác giả lưu vào author_list
author_list = set()
b = []
for i in df_author['author']:
    s = str(i).split('|')
    for j in s:
        author_list.add(j)
```

```python
author_list
```

<h3>Bước 2:</h3> Tạo tập vector đặc trưng cho mỗi bài báo, mỗi vectơ đặc trưng là one-hot vectơ với mỗi vị trí phần tử đại diện cho thứ tự bài báo, ở đây ta sẽ lưu vào bảng df_pp mỗi cột sẽ là một one-hot vector của bài báo<p>

```python
# thêm cột idpp cho df_author (cột này đơn giản để đánh số thứ tự cho các bài báo)
df_author['idpp'] = np.arange(df_author.shape[0])
```

```python
# cột idpp đã được thêm vào 
df_author.head()
```

```python
#Các one-host vector của các bài báo sẽ lưu vào df_pp, mỗi cột là một bài báo
df_pp = pd.get_dummies(df_author.loc[:,'idpp'], prefix='pp')
```

```python
df_pp.head()
```

<h3>Bước 3:</h3> Tạo tập hồ sơ người dùng: Mỗi nghiên cứu viên có vectơ là tổng các one-hot vectơ của các bài báo được công bố bởi nghiên cứu viên ấy.<p>

```python
# khởi tạo tập hồ sơ người dùng, mỗi hàng sẽ là một one-hots vector của một tác giả
author_vt = np.zeros((len(author_list), df_author.shape[0]))
author_vt = pd.DataFrame(author_vt, index = list(author_list))
```

```python
# hiện tại author_vt mới được khởi tạo và tất cả các giá trị đều là 0
author_vt.head()
```

```python
# Vòng for đầu tiên để lặp qua từng bài báo
# Vòng for thứ 2 để kiểm tra với bài báo thứ i thì có những tác giả nào, sau đó ta sẽ cộng thêm 1 vào vị trí thứ i của từng vector của từng tác giả có trong bài báo i
for i in range(df_author.shape[0]):
    str_author= str(df_author['author'][i]).split('|')
    for auth in str_author:
        author_vt.loc[auth,i] = author_vt.loc[auth,i]  + 1
```

```python
# sau khi chạy xong ta sẽ có một author_vt hoàn chỉnh, với mỗi cột là vector đặc trưng của từng tác giả
author_vt = author_vt.transpose()
author_vt.head()
```

<h3>Bước 4:</h3> Khi tìm kiếm một người cộng tác với ứng viên A, ta sẽ tính độ tương tự cosine của ứng viên A này với các ứng viên khác dựa trên "tập hồ sơ người dùng" tính được ở bước 3


```python
from sklearn.metrics.pairwise import cosine_similarity

# trong đoạn code bên dưới là ta đang tính độ tương tự cosine của tác giả "Tram Ngoc Pham" với từng tác giả có trong bảng author_vt
# Công thức tính: https://en.wikipedia.org/wiki/Cosine_similarity
# sau khi tính xong ta sẽ có top 10 ứng viên tương tự nhất với "Tram Ngoc Pham" lưu vào biến df_cosine
cosine = {}
for author in author_vt.columns:
    cosi = cosine_similarity(author_vt[["Tram Ngoc Pham"]].values.reshape(1, -1), author_vt[[author]].values.reshape(1, -1))
    cosine[author] = cosi[0]
    
df_cosine = pd.DataFrame(cosine, index = ['cosine_similarity'])
df_cosine = df_cosine.transpose()
df_cosine.sort_values(by=['cosine_similarity'], ascending = False, inplace= True)
df_cosine.drop(["Tram Ngoc Pham"], inplace = True)
```
